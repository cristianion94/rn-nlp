```python
import wikipedia
```

```python
wikipedia.set_lang("ro")
```

```python
topics = ["romania", "berlin"]
```

```python
corpus = []
```

```python
def search(topic):
    pages = wikipedia.search(topic)
    for page in pages:
        content = wikipedia.summary(page)
        corpus.append(content)
```

```python
for topic in topics:
    search(topic)
```

```python
N = 2
laplaceSmoothingConstant = 100

apparitions = dict()


# previouslyComputedProbability = dict()

def computeApparitionsCounter(corpus):
    global laplaceSmoothingConstant
    uniqueWords = set()
    for sentence in corpus:
        words = sentence.split(' ')
        words = ["<s>"] + words + ["<\s>"]

        previousWordsList = list()
        for i in range(N - 1):
            uniqueWords.add(words[i])
            previousWordsList.append(words[i])
        for i in range(N - 1, len(words) + 1):
            previousWords = ' '.join(previousWordsList)
            if previousWords not in apparitions:
                apparitions[previousWords] = 0
            apparitions[previousWords] += 1
            if (i < len(words)):
                uniqueWords.add(words[i])
                fullWords = previousWords + ' ' + words[i]
                if fullWords not in apparitions:
                    apparitions[fullWords] = 0
                apparitions[fullWords] += 1

                previousWordsList.append(words[i])
                previousWordsList.pop(0)


    laplaceSmoothingConstant = len(uniqueWords)


def computeWordNewProbability(previousWords, word):
    # if frozenset((previousWords, word)) in previouslyComputedProbability:
    #    return previouslyComputedProbability[frozenset((previousWords, word))]
    nrApparitionsFull = 1
    nrApparitionsPrevious = laplaceSmoothingConstant
    fullWords = previousWords + " " + word
    if fullWords in apparitions:
        nrApparitionsFull += apparitions[fullWords]
    if previousWords in apparitions:
        nrApparitionsPrevious += apparitions[previousWords]
    # previouslyComputedProbability[frozenset((previousWords, word))] = nrApparitionsFull / nrApparitionsPrevious
    print("{1} | {0} : {2} / {3}".format(previousWords, word, nrApparitionsFull, nrApparitionsPrevious))
    return nrApparitionsFull / nrApparitionsPrevious


def computeProbability(s):
    result = 1
    words = s.split(' ')
    words = ["<s>"] + words + ["<\s>"]
    for i in range(N - 1, len(words)):
        previousWords = ""
        for i1 in range(N - 1, 0, -1):
            previousWords += words[i - i1] + " "
        previousWords = previousWords[:-1]
        print(previousWords)
        tmp = computeWordNewProbability(previousWords, words[i])
        result *= tmp
    return result


N = 3
computeApparitionsCounter(corpus)
print(computeProbability("pe cursul inferior al Dunării este o casa"))
print(computeProbability("pe cursul inferior al al este o casa"))
```
